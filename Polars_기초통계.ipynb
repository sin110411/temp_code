```python
import polars as pl
from datetime import datetime

#폴라스로 데이터 프레임 만들기
temp_df = pl.DataFrame({
    'integer' : [1,2,3],
    'date' : [
        datetime(2025,1,1),
        datetime(2025,1,2),
        datetime(2025,1,3)
    ],
    'float' : [4.0, 5.0, 6.0],
    'string' : ['a', 'b', 'c']
})

print(temp_df) #폴라스로 만든 데이터프레임은 주피터에서 미리보기도 지원하지만 
# print()를 사용하면 아래와같은 형태로 볼수 있음


#해당경로에 폴라스df저장하기
temp_df.write_csv('폴라스연습_df.csv')#폴라스에는 판다스처럼 encoding적지않음

#저장한 폴라스df불러오기
temp_df1=pl.read_csv('폴라스연습_df.csv')

print(temp_df)
```

```python
#polars의 select
#데이터프레임에서 열을 선택하려면 두가지 작업수행
# 1. 얻고자 하는 데이터를 원하는 DataFrame에 정의
# 2. 기존 DataFrame에서 필요한 데이터 선택하기


#단순 selelct만 사용하면 단순 컬럼자체만을가져오기때문에 컬럼을 가지고 연산 불가능(단순 확인용)
# pl.col로 데이터를 가져오면 연산이 가능한 형태로 데이터를 가져옴


#단순 select 사용
print(temp_df.select('*')) #모든 데이터불러오기(SQL과 동일문법)
print(temp_df.select('date'))
print(temp_df.select('integer'))
print(temp_df.select('integer','string'))
print()
print("======================================================")
print()


# pl.col()사용
print(temp_df.select(pl.col('*')))
print(temp_df.select(pl.col('integer','string')))
print(temp_df.select(pl.col('*').exclude('date'))) #pl.col('*').exclude('해당열') exclude에 정의된 열 말고 전부 가져오기

out = ['intager', 'date']
print(temp_df.select(pl.col('*').exclude(out))) # 변수에 저장해서도 사용가능!
```

```python
#polars의 filter
# filter옵션을 사용하면 DataFrame의 하위 집합만들기 가능
# filter를 걸 열을 pl.col로 선택하고 조건문을 통해 생성하는것이기때문에 filter안에 pl.col이 들어가는 형태로 작성

print(temp_df.filter(pl.col('integer')>=2)) #integer의 열에서 2이상인것만 가져오기
print(temp_df.filter(pl.col('date').is_between(datetime(2025,1,1), datetime(2025,1,2))))

# and,or,not조건도 사용가능. 하지만 각각의 조건문을 filter()괄호안에서 각각 괄호로 묶어주어야함
# polar에서 조건문 쓸때 연산자 우선순위를 잘 지켜주어야함
# &(and)는 ==,<=,=>보다 연산순위가 낮아서 문제가 생기지 않기때문에 첫번째 조건문은 괄호로 감싸주지 않았지만
# 두번째 |의 경우 ==이 |보다 낮거나하지않기때문에 파이썬에서 자동으로 'b'| 의 뭔가를 먼저 시도해버려서 에러나기때문에
# |와 ==를 쓸때는 괄호롤 감싸주어야함

#헷갈리면 조건문자체는 각각 개별로 그냥 괄호로 감싸주기!
print(temp_df.filter(pl.col('float').is_not_nan() & (pl.col('integer')==1)))
print(temp_df.filter((pl.col('float').is_not_nan()) & (pl.col('integer')==1)))
print(temp_df.filter((pl.col('string')=='b') | (pl.col('float')>=3.0)))
```

```python
# with_columns 
# 새로운 열 추가
# 분석하고 난 결과물을 새로운 열에 추가해야할때 등등 사용
# alisa(열이름)과 같이 사용하는것이 좋음


print(temp_df.with_columns((pl.col('integer')+10).alias('ten')))
print(temp_df.with_columns(pl.col('integer').sum().alias('intsum')))
print(temp_df.with_columns((pl.col('string')+'aa').alias('add_str')))
```

```python
# group_by
# 판다스의 groupby와 동일 기능

#새로운 데이터프레임 생성

temp_df1 = pl.DataFrame({
    'x' : range(8),
    'y' : ['A','A','A','B','B','C','X','X']
})

print(temp_df1)
```

```python
#maintain_order = True : 폴라스에서 그룹을 하면 자동정렬이되는데 자동정렬하지말고 
# 그룹하기 전 기존 열의 순서를 동일하게 유지하라는 뜻
print(temp_df1.group_by('y', maintain_order = True).len())
print(temp_df1.group_by('y',maintain_order = True).sum())
print(temp_df1.group_by('y',maintain_order = True).mean())


print()
print('=============================')
print()


# agg
# 판다스의 agg와 동일기능. agg를 사용하면 그룹한 결과에 함수를 사용할 수 있음
print(temp_df1.group_by('y', maintain_order = True).agg(
    pl.col('*').count().alias('count'),
    pl.col('*').sum().alias('sum')
))

print(temp_df1.group_by('y',maintain_order = True).agg(
    pl.col('*').mean().alias('평균'),
    pl.col('*').sum().alias('합계')
))
```

```python
# Polars 실습

# 방문데이터 생성
data = {
    "visit_id": [f"V{str(i).zfill(5)}" for i in range(1, 31)],
    "customer_id": [
        "C1001", "C1002", "C1001", "C1003", "C1002", "C1004",
        "C1005", "C1006", "C1001", "C1007", "C1008", "C1003",
        "C1009", "C1005", "C1010", "C1008", "C1011", "C1012",
        "C1001", "C1002", "C1013", "C1010", "C1014", "C1015",
        "C1006", "C1003", "C1004", "C1002", "C1005", "C1016"
    ],
    "visit_date": [
        "2024-06-01", "2024-06-01", "2024-06-10", "2024-06-05", "2024-06-12", "2024-06-07",
        "2024-06-09", "2024-06-03", "2024-06-15", "2024-06-20", "2024-06-18", "2024-06-22",
        "2024-06-21", "2024-06-25", "2024-06-28", "2024-06-29", "2024-07-01", "2024-07-01",
        "2024-07-02", "2024-07-03", "2024-07-05", "2024-07-07", "2024-07-08", "2024-07-09",
        "2024-07-10", "2024-07-11", "2024-07-12", "2024-07-13", "2024-07-14", "2024-07-15"
    ],
    "gender": [
        "M", "F", "M", "M", "F", "F",
        "M", "F", "M", "F", "F", "M",
        "F", "M", "M", "F", "F", "M",
        "M", "F", "F", "M", "M", "F",
        "F", "M", "F", "F", "M", "F"
    ],
    "age": [
        30, 25, 30, 42, 25, 35,
        30, 20, 30, 45, 33, 42,
        29, 30, 55, 33, 22, 39,
        30, 25, 40, 55, 28, 37,
        20, 42, 35, 25, 30, 31
    ],
    "amount": [
        5200, 12000, 7500, 3000, 9000, 11000,
        4000, 8000, 3000, 10000, 7000, 3500,
        12000, 6000, 4000, 3000, 9500, 4700,
        7000, 8000, 10000, 4500, 8200, 9800,
        7500, 4200, 6600, 7300, 5500, 8700
    ]
}

# 데이터프레임 생성
df = pl.DataFrame(data)

print(df)
```

```python
# 실습목록

# 1	나이별 평균, 합계 소비금액	
# 2	성별별 평균 소비금액 비교	
# 3	고객별 총 방문 횟수	
# 4	고객별 총 소비 금액	
# 5	재방문 고객만 추출
# 6	전체 재방문율 = 재방문 고객 수 / 전체 고객 수	비교용 연산
# 7	월별 소비 금액 합계	
# 8	1인당 평균 소비액이 높은 연령대 상위 3개	
```

```python
# 1	나이별 평균 소비금액	


#나이 정규화
#cast : 데이터타입 변경 및 설정할때 사용
df = df.with_columns((pl.col('age')//10).cast(pl.Int64).alias('나이 정규화'))
df

print(df.group_by('나이 정규화', maintain_order = True).agg(
    pl.col('amount').mean().alias('나이별 평균 소비금액'),
    pl.col('amount').sum().alias('나이별 합계 소비금액')
))
```

```python
# 2	성별별 평균 소비금액 비교	

print(df.group_by('gender',maintain_order = True).agg(
    pl.col('amount').mean().alias('성별별 평균 소비금액')
))
```

```python
# 3	고객별 총 방문 횟수	

print(df.group_by('customer_id',maintain_order = True).agg(
    pl.col('visit_id').count().alias('고객별 총 방문 횟수')
))

# + 총 고객수 구하기 (행,열 개수 참고) 16명
print(df.group_by('customer_id',maintain_order = True).count())
```

```python
# 5	재방문 고객만 추출

# 그룹바이한 상태에서 그룹된 결과물에 조건을 걸고싶다면 agg로 함수문안에서 에시컬럼(alias)을 생성한 후 
# 괄호로 닫아 끝내고 그 괄호에 덧붙여 filter를 사용하여 예시컬럼에 조건문을 걸어야함 
print(df.group_by('customer_id',maintain_order = True).agg(
    pl.col('visit_id').count().alias('고객별 방문횟수')).filter(
        pl.col('고객별 방문횟수')>=2))

# + 총 사용금액 합계가 50000만원 이상인 나이대만 추출
print(df.group_by('나이 정규화',maintain_order = True).agg(
    pl.col('amount').sum().alias('나이별 합계금액')).filter(
    pl.col('나이별 합계금액')>=50000
    ))
```

```python
# 6	전체 재방문율 구하기

# 전체 수, 방문자수 등등은 굳이 구하지 않아도 조건에 맞게 데이터를 추출하면 행,열 수만 봐도 된다는것 기억하기
#전체방문자수 : 16명
# 재 방문자수 : 8명
print(f'재 방문률 : {8/16*100}%')
```

```python
# 7	월별 소비 금액 합계

#폴라스에서 문자열을 다룰때는 반드시 .str.기타 다른 함수의 형식으로 사용
# 문자열을 다루기 위해 str을 붙여서 사용하고 문자열 슬라이싱을 할대는 str.slice(시작,끝)을 파라미터값으로 주어야함
# 슬라이스는 선택한 열에 대해 실행하는 것이기 때문에 반드시 pl.col에 붙여서 사용
# 폴라스의 경우 슬라이싱 끝지점보다 덜 가져오는거 없이 입력한 숫자 그대로 가져옴

df = df.with_columns(pl.col('visit_date').str.slice(0,7).alias('년월'))
df


print(df.group_by('년월',maintain_order = True).agg(
    pl.col('amount').sum().alias('년월별 소비액 합계')
))
```

```python
# 8 평균 소비액이 높은 연령대 상위 3개

#오름차순, 내립차순 하려면 sort('컬렴명', descending = True / False)로 지정
#원하는 개수만큼 추출할때는 head(n)사용

print(df.group_by('나이 정규화',maintain_order = True).agg(
    pl.col('amount').mean().alias('연령대별 평균 소비금액')).sort(
    pl.col('연령대별 평균 소비금액'), descending = True).head(3))
```
